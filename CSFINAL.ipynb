```python
#turnover calculation below 1Cr
def turnoverb100(A,B,C,D,E,G):
    count = 0
    if(A>=G):
        count +=1
    if(B>=G):
        count +=1
    if(C>=G):
        count +=1
    if(D>=G):
        count+=1
    if(E>=G):
        count+=1
    if(count >= 2):
        return 1
    return 0 

#turnover calculation below 1Cr
def turnovera100(A,B,C,D,E,G):
    count = 0
    if(A>=(2*G)):
        count +=1
    if(B>=(2*G)):
        count +=1
    if(C>=(2*G)):
        count +=1
    if(D>=(2*G)):
        count+=1
    if(E>=(2*G)):
        count+=1
    if(count >= 1):
        return 1
    return 0 
def lm(A,B):
    if(A>=B):
        return "more"
    return "less"

def st(A,B):
    if(A>=B):
        return "Satisfies"
    return "Does not satisfy"

def TB(A,B,C,D,E,G):
    count = 0
    if(A>=G):
        count +=1
    if(B>=G):
        count +=1
    if(C>=G):
        count +=1
    if(D>=G):
        count+=1
    if(E>=G):
        count+=1
    if(count >= 2):
        return "Satisfies"
    return "Does not satisfy" 

#turnover calculation below 1Cr
def TO(A,B,C,D,E,G):
    count = 0
    if(A>=(2*G)):
        count +=1
    if(B>=(2*G)):
        count +=1
    if(C>=(2*G)):
        count +=1
    if(D>=(2*G)):
        count+=1
    if(E>=(2*G)):
        count+=1
    if(count >= 1):
        return "Satisfies"
    return "Does not satisfy" 
```

```python
#calculation maximum of turnovers
def maxi(A,B,C,D,E):
    maxiu= A
    if(B>maxiu):
        maxiu = B
    if(C>maxiu):
        maxiu = C
    if(D>maxiu):
        maxiu = D
    if(E>maxiu):
        maxiu = E
    return maxiu
```
